```python
# Feature Scaling

sc = MinMaxScaler(feature_range = (0, 1))
training_set_scaled_ambev = sc.fit_transform(set_treino_ambev)
training_set_scaled_itau = sc.fit_transform(set_treino_itau)
training_set_scaled_petrobras = sc.fit_transform(set_treino_petrobras)
training_set_scaled_gol = sc.fit_transform(set_treino_gol)

```

```python
# Creating a data structure with 60 time-steps and 1 output
# CRIADO APENAS PARA AMBEV
# CRIADO APENAS PARA GOL
# CRIADO APENAS PARA PETROBRAS
# CRIADO APENAS PARA ITAU


X_train = []
y_train = []

for i in range(60, 800):
    X_train.append(training_set_scaled_ambev[i-60:i, 0])
    y_train.append(training_set_scaled_ambev[i, 0])
X_train, y_train = np.array(X_train), np.array(y_train)
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
```

```python
model = Sequential()
#Adding the first LSTM layer and some Dropout regularisation
model.add(LSTM(units = 50, return_sequences = True, input_shape = (X_train.shape[1], 1)))
model.add(Dropout(0.3))
# Adding a second LSTM layer and some Dropout regularisation
model.add(LSTM(units = 50, return_sequences = True))
model.add(Dropout(0.3))
# Adding a third LSTM layer and some Dropout regularisation
model.add(LSTM(units = 50, return_sequences = True))
model.add(Dropout(0.3))
# Adding a fourth LSTM layer and some Dropout regularisation
model.add(LSTM(units = 50))
model.add(Dropout(0.3))
# Adding the output layer
model.add(Dense(units = 1))

# Compiling the RNN
model.compile(optimizer = 'adam', loss = 'mean_squared_error', metrics=['mse', 'mae', 'mape','accuracy'])

# Fitting the RNN to the Training set
history = model.fit(X_train, y_train,  epochs = 1000, batch_size = 128)
pyplot.plot(history.history['mse'], color='blue', label='MSE')
pyplot.plot(history.history['mae'], color='red', label='MAE')
pyplot.plot(history.history['mape'], color='green', label='MAPE')
plt.title('Convergência do modelo')
plt.xlabel('Épocas')
plt.ylabel('Perda')
plt.legend()
pyplot.show()
```

```python
# predict and plot function
def predict_prices(prices, x):
    dates = np.reshape(dates,(len(dates), 1)) # convert to 1xn dimension
    x = np.reshape(x,(len(x), 1))
    
    #svr_lin  = SVR(kernel='linear', C=1e3)
    #svr_poly = SVR(kernel='poly', C=1e3, degree=2)
    svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.1)
    
    # Fit regression model
    #svr_lin .fit(dates, prices)
    #svr_poly.fit(dates, prices)
    svr_rbf.fit(y, prices)
    
    plt.scatter(y, prices, c='k', label='Data')
    #plt.plot(dates, svr_lin.predict(dates), c='g', label='Linear model')
    plt.plot(dates, svr_rbf.predict(y), c='r', label='RBF model')    
    #plt.plot(dates, svr_poly.predict(dates), c='b', label='Polynomial model')
    
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title('Support Vector Regression')
    plt.legend()
    plt.show()
    
    return svr_rbf.predict(x)[0]
```
